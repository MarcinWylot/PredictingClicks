```python
numerical = ['express.no.transactions', 'express.total.spend','metro.no.transactions','metro.total.spend', \
             'superstore.no.transactions','superstore.total.spend','extra.no.transactions','extra.total.spend',\
             'fandf.no.transactions','fandf.total.spend','petrol.no.transactions','petrol.total.spend', \
             'direct.no.transactions','direct.total.spend']

categorical = ['gender','affluency','county']

content_list = ['content_1','content_2', 'content_3','content_4','content_5','content_6', \
                'content_7','content_8','content_9']

data_file = r'C:\Users\martinX1\Dropbox\projects\PredictingClicks\content_train.tsv'
```

```python
import csv
import pandas
import numpy

def bulild_features(file,numerical,categorical,content_list):
    with open(file, "r") as f:
        reader = csv.DictReader(f, delimiter='\t') #for performance I don't work with DataFrame here
        new_data = [] #it's much faster to use list of dictionaries to construc a new dataset
        for row in reader:
            for content in content_list:
                if content in row and row[content] != 'NA':
                    row_new = {_:row[_] for _ in numerical+categorical}
                    row_new['content'] =content
                    row_new['click'] = int(row[content])
                    new_data.append(row_new)
        df = pandas.DataFrame(new_data)
        df[numerical] = df[numerical].astype(numpy.float64)
        return df          
 
    
%time new_data = bulild_features(data_file, numerical, categorical, content_list)
new_data.head()
```
